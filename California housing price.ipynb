## Regression Models

### Linear regression

```python
from sklearn.linear_model import LinearRegression

linear_req=LinearRegression()
linear_req.fit(train_x,train_y)
```

```python
from sklearn.model_selection import cross_val_score 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Predicting Cross Validation Score the Test set results
cv_linear=cross_val_score(estimator=linear_req,X=train_x,y=train_y,cv=10)

# predict R2 for trainng data set
linear_train_prediction=linear_req.predict(train_x)
r2_score_train_linear=r2_score(train_y,linear_train_prediction)

# predict R2 for test data set
linear_test_prediction=linear_req.predict(test_x)
r2_score_test_linear=r2_score(test_y,linear_test_prediction)

# predict Mean_Square_Error for test data set
mse=mean_squared_error(test_y,linear_test_prediction)

# predict RMSE for test data set
rmse_linear=mean_squared_error(test_y,linear_test_prediction)
print("Cross Validation_linear: ", cv_linear.mean())
print("R2_score_linear_train: ", r2_score_train_linear)
print("R2_score_linear_test: ",r2_score_test_linear)
print("Mean_Squared_Error_linear: ",mse)
print("Root_Mean_Squared_Error_linear: ",rmse_linear)
```

### Polynomial Regression - 2nd degree Â¶

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(train_x)
poly_reg.fit(X_poly, train_y)
regressor_poly2 = LinearRegression()
regressor_poly2.fit(X_poly, train_y)
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.metrics import mean_squared_error

# Predicting Cross Validation Score the Test set results
cv_poly2 = cross_val_score(estimator = regressor_poly2, X = train_x, y = train_y, cv = 10)

# Predicting R2 Score the Train set results
y_pred_poly2_train = regressor_poly2.predict(poly_reg.fit_transform(train_x))
r2_score_poly2_train = r2_score(train_y, y_pred_poly2_train)

# Predicting R2 Score the Test set results
y_pred_poly2_test = regressor_poly2.predict(poly_reg.fit_transform(test_x))
r2_score_poly2_test = r2_score(test_y, y_pred_poly2_test)

# Predicting RMSE the Test set results
rmse_poly2 = (np.sqrt(mean_squared_error(test_y, y_pred_poly2_test)))
print('Cross Validation_poly2: ', cv_poly2.mean())
print('R2_score_poly2_train: ', r2_score_poly2_train)
print('R2_score_poly2_test: ', r2_score_poly2_test)
print("Root_Mean_Squared_Error_poly2: ", rmse_poly2)
```

### Ridge Regression Â¶

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

steps = [
    ('scalar', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)),
    ('model', Ridge(alpha=3.8, fit_intercept=True))
]

ridge_pipe = Pipeline(steps)
ridge_pipe.fit(train_x, train_y)
```

```python
# Predicting Cross Validation Score the Test set results
cv_ridge = cross_val_score(estimator = ridge_pipe, X = train_x, y = train_y.ravel(), cv = 10)

# Predicting R2 Score the Test set results
y_pred_ridge_train = ridge_pipe.predict(train_x)
r2_score_ridge_train = r2_score(train_y, y_pred_ridge_train)

# Predicting R2 Score the Test set results
y_pred_ridge_test = ridge_pipe.predict(test_x)
r2_score_ridge_test = r2_score(test_y, y_pred_ridge_test)

# Predicting RMSE the Test set results
rmse_ridge = (np.sqrt(mean_squared_error(test_y, y_pred_ridge_test)))
print('Cross Validation_ridge: ', cv_ridge.mean())
print('R2_score_ridge_train: ', r2_score_ridge_train)
print('R2_score_ridge_test: ', r2_score_ridge_test)
print("Root_Mean_Squared_Error_ridge: ", rmse_ridge)
```

### Lasso Regression Â¶

```python
from sklearn.linear_model import Lasso

steps = [
    ('scalar', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)),
    ('model', Lasso(alpha=0.012, fit_intercept=True, max_iter=3000))
]

lasso_pipe = Pipeline(steps)
lasso_pipe.fit(train_x,train_y)
```

```python
# Predicting Cross Validation Score the Test set results
cv_lasso = cross_val_score(estimator = ridge_pipe, X= train_x, y = train_y.ravel(), cv = 10)

# Predicting R2 Score the Test set results
y_pred_lasso_train = lasso_pipe.predict(train_x)
r2_score_lasso_train = r2_score(train_y, y_pred_lasso_train)

# Predicting R2 Score the Test set results
y_pred_lasso_test = lasso_pipe.predict(test_x)
r2_score_lasso_test = r2_score(test_y, y_pred_lasso_test)

# Predicting RMSE the Test set results
rmse_lasso = (np.sqrt(mean_squared_error(test_y, y_pred_lasso_test)))
print('Cross Validation:_lasso ', cv_lasso.mean())
print('R2_score_lasso_train: ', r2_score_lasso_train)
print('R2_score_lasso_test: ', r2_score_lasso_test)
print("Root_Mean_Squared_Error_lasso: ", rmse_lasso)
```

### Suport Vector Regression 

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaling_X = StandardScaler()
scaling_y = StandardScaler()
scaled_x = scaling_X.fit_transform(train_x)
scaled_y = scaling_y.fit_transform(train_y.values.reshape(-1,1))
```

```python
# Fitting the Suport vector Regression Model to the dataset
from sklearn.svm import SVR
SVR_reg = SVR(kernel = 'rbf', gamma = 'scale')
SVR_reg.fit(scaled_x, scaled_y.ravel())
```

```python
# Predicting Cross Validation Score the Test set results
cv_SVR = cross_val_score(estimator = SVR_reg, X= scaled_x, y = scaled_y.ravel(), cv = 10)

# Predicting R2 Score the Test set results
y_pred_svr_train = scaling_y.inverse_transform(SVR_reg.predict(scaling_X.transform(train_x)))
r2_score_cvr_train = r2_score(train_y, y_pred_svr_train)

# Predicting R2 Score the Test set results
y_pred_svr_test = scaling_y.inverse_transform(SVR_reg.predict(scaling_X.transform(test_x)))
r2_score_cvr_test = r2_score(test_y, y_pred_svr_test)

# Predicting RMSE the Test set results
rmse_svr = (np.sqrt(mean_squared_error(test_y, y_pred_svr_test)))
print('Cross Validation_svr: ', cv_SVR.mean())
print('R2_score_svr_train: ', r2_score_cvr_train)
print('R2_score_svr_test: ', r2_score_cvr_test)
print("Root_Mean_Squared_Error_svr: ", rmse_cvr)
```

### Decision Tree Regression Â¶


```python
# Fitting the Decision Tree Regression Model to the dataset
from sklearn.tree import DecisionTreeRegressor
dec_tree_req=DecisionTreeRegressor(random_state = 0)
dec_tree_req.fit(train_x,train_y)
```

```python
# Predicting Cross Validation Score the Test set results
cv_dec_tree = cross_val_score(estimator = dec_tree_req, X= train_x, y = train_y, cv = 10)

# Predicting R2 Score the Test set results
y_pred_dec_tree_train = dec_tree_req.predict(train_x)
r2_score_dec_tree_train = r2_score(train_y, y_pred_dec_tree_train)

# Predicting R2 Score the Test set results
y_pred_dec_tree_test = dec_tree_req.predict(test_x)
r2_score_dec_tree_test = r2_score(test_y, y_pred_dec_tree_test)

# Predicting RMSE the Test set results
rmse_dec_tree = (np.sqrt(mean_squared_error(test_y, y_pred_dec_tree_test)))
print('Cross Validation_dec_tree: ', cv_dec_tree.mean())
print('R2_score_dec_tree_train: ', r2_score_cvr_train)
print('R2_score_dec_tree_test: ', r2_score_cvr_test)
print("Root_Mean_Squared_Error_dec_tree: ", rmse_dec_tree)
```

### Random Forest Regression Â¶

```python
#fitting the Random Forest Regression Model 
from sklearn.ensemble import RandomForestRegressor
rand_forest_req=RandomForestRegressor(n_estimators = 500, random_state = 0)
rand_forest_req.fit(train_x,train_y)
```

```python
# Predicting Cross Validation Score the Test set results
cv_d_rand_forest = cross_val_score(estimator = rand_forest_req, X= scaled_x, y = train_y.ravel(), cv = 10)

# Predicting R2 Score the Test set results
y_pred_rand_forest_train = rand_forest_req.predict(train_x)
r2_score_rand_forest_train = r2_score(train_y, y_pred_rand_forest_train)

# Predicting R2 Score the Test set results
y_pred_rand_forest_test = rand_forest_req.predict(test_x)
r2_score_rand_forest_test = r2_score(test_y, y_pred_rand_forest_test)

# Predicting RMSE the Test set results
rmse_rand_forest = (np.sqrt(mean_squared_error(test_y, y_pred_rand_forest_test)))
print('Cross Validation_rand_forest: ', cv_d_rand_forest.mean())
print('R2_score_rand_forest_train: ', r2_score_rand_forest_train)
print('R2_score_rand_forest_test: ', r2_score_rand_forest_test)
print("Root_Mean_Squared_Error_rand_forest: ", rmse_rand_forest)
```
