### Balancing Dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import ExtraTreesClassifier  
from imblearn.over_sampling import SMOTE
```

```python
X=data.drop(['Customer attrition'],axis=1)
y=data['Customer attrition']
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.3,random_state=0)
sb.countplot(y)
```

```python
train_y
```

```python
smote=SMOTE(sampling_strategy='minority', k_neighbors=20, random_state=42)
resample_train_x, resample_train_y = smote.fit_resample(train_x, train_y)
sb.countplot(resample_train_y)
```

### Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

random_model=RandomForestClassifier()
random_model.fit(train_x,train_y)
random_pred_model=random_model.predict(test_x)


print("mean_squared_error:",mean_squared_error(test_y,random_pred_model))
print("accuracy_score:",accuracy_score(test_y,random_pred_model))
print(confusion_matrix(test_y,random_pred_model))
print(classification_report(test_y,random_pred_model))

plot_confusion_matrix(random_model, test_x, test_y)
plt.show()
```

### Support Vector Machine


```python
from sklearn.svm import SVC 
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import plot_confusion_matrix

model_svm = make_pipeline(StandardScaler(), SVC(gamma='auto'))
model_svm.fit(resample_train_x, resample_train_y)
svm_pred_model= model_svm.predict(test_x)

print("mean_squared_error:",mean_squared_error(test_y,svm_pred_model))
print("accuracy_score:",accuracy_score(test_y,svm_pred_model))
print(confusion_matrix(test_y,svm_pred_model))
print(classification_report(test_y,svm_pred_model))

plot_confusion_matrix(model_svm, test_x, test_y)
plt.show()
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gr_b_model=GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=42)
gr_b_model.fit(resample_train_x, resample_train_y)
gr_b_pred= gr_b_model.predict(test_x)

print("mean_squared_error:",mean_squared_error(test_y,gr_b_pred))
print("accuracy_score:",accuracy_score(test_y,gr_b_pred))
print(confusion_matrix(test_y,gr_b_pred))
print(classification_report(test_y,gr_b_pred))

plot_confusion_matrix(gr_b_model, test_x, test_y)
plt.show()
```
